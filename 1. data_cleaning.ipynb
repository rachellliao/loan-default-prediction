## Cleaning

### Handling Missing Value

```python
# Check the percentage of missing values in each column
df.isnull().sum() / len(df) * 100
```

```python
# Create a copy of the original dataframe
df_clean = df.copy()

# Get the columns with more than 20% missing values
cols_to_drop = df_clean.columns[df_clean.isnull().sum() / len(df_clean) * 100 > 20]

# Drop the columns with more than 20% missing values
df_clean.drop(cols_to_drop, axis=1, inplace=True)
```

```python
# Check the shape of the dataframe
df_clean.shape
```

```python
# Check the percentage of missing values in each column
df_clean.isnull().sum() / len(df_clean) * 100
```

Let's investigate the columns with relatively high missing values comparing to the rest of the dataset:
1. il_util (~17% of the rows are missing)

2. mths_since_recent_inq (~10% of the rows are missing)

```python
# Exam the il_util column
col_info('il_util')
```

**Interpretation of the column:**

The column "il_util" refers to a financial metric that calculates the percentage of the total current balance owed on all installment accounts in relation to the high credit or credit limit allowed on those accounts.

**Timing of Variable in the Loan Application Process:**

Arises **after** Loan Disbursement.

Information about the current balances on installment accounts is determined after the loan has been disbursed and the borrower starts using their installment accounts.

**Decision on data handling:**

Since it arises after loan disbursement, it is not useful for loan prediction analysis, let's **drop** this column.


```python
# drop the il_util column
df_clean.drop('il_util', axis=1, inplace=True)
```

```python
# Exam the mths_since_recent_inq column
col_info('mths_since_recent_inq')
```

**Interpretation of the column:**

The "mths_since_recent_inq" column in the LendingClub loan data tells us the time that has passed since the last hard credit inquiry was made on the borrower's credit report at the time of the loan application. 

It provides information about how actively the borrower has been seeking credit recently, and it can be a factor considered by LendingClub when evaluating a loan application. Generally, a longer period since the last inquiry is considered more favorable by lenders, as it suggests that the borrower is not taking on new credit obligations too frequently.

**Timing of Variable in the Loan Application Process:**

Arises **before** Loan Disbursement.

Credit inquiries occur when LendingClub check the borrower's credit report as part of the loan application process. It reflects the borrower's recent credit activity and credit-seeking behavior at the time of applying for the loan.

**Decision on data handling:**

Since it arises before loan disbursement, it is useful for loan prediction analysis, let's **keep** this column.

```python
# Get the columns with less than 20% missing values
cols_to_fill = df_clean.columns[df_clean.isnull().sum() / len(df_clean) * 100 < 20]

# Fill the missing values with the mean of the respective columns for float columns
for col in cols_to_fill:
    if df_clean[col].dtype == 'float64':
        df_clean[col].fillna(df_clean[col].mean(), inplace=True)

# Check the percentage of missing values in each column 
df_clean.isnull().sum() / len(df_clean) * 100
```

Let's investigate the columns with relatively high missing values comparing to the rest of the dataset:
1. emp_title (~6% of the rows are missing)

2. emp_length (~6% of the rows are missing)
